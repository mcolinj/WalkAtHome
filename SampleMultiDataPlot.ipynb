```python
from matplotlib.pylab import plt
import steps
s = steps.Steps(26, "RPD-WalkAtHome-20200424.csv")

def multi_plot(s, steppers):
    patterns = [ '-w', '-v', '-x', '-.', '--',  '-v', '-x' ]
    colors = [ "orange", "maroon", "blue", "black", 'chocolate', 'green', 'red' ]  
    for idx,stepper in enumerate(steppers):
        plt.plot(s.__getattr__(stepper), patterns[idx], color=colors[idx], label=stepper)
    
    plt.legend(loc='lower right')
    plt.show() 

%matplotlib inline

multi_plot( s, ("We", "Bo", "Yi", "Co", "Pa", "Vr", "Ki") )






```
